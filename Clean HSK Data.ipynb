```python
df = pd.read_csv('hsk30.csv')
```

```python
df = df[['Simplified', 'Traditional', 'Pinyin', 'Level']]
```

```python
df.sample(50)
```

```python
def clean_string(string):
    if re.search('\（.*\）$', string):
        return string[:string.index('（')]
    if re.search('\(.*\)$', string):
        return string[:string.index('(')].strip()
    if re.search('\（.*\）', string):
        string1 = re.sub('\（.*\）', '', string)
        string2 = string.replace('（', '').replace('）', '')
        return (string1, string2)
    if re.search('\(.*\)', string):
        string1 = re.sub('\(.*\)', '', string)
        string2 = string.replace('(', '').replace(')', '')
        return (string1, string2)
    if re.search('\|', string):
        string1 = string.split('|')[0]
        string2 = string.split('|')[1]
        return (string1, string2)
    return string
```

```python
# some words have definitions like 差(一)点儿 which should really be split into 差点儿 and 差一点儿
# other words that have parentheses to show an example of usage need those parentheses removed

cleaned_dict = {'Simplified': [], 'Traditional': [], 'Pinyin': [], 'Level': []}

def split_row(row):
    simplified_string = clean_string(row['Simplified'])
    traditional_string = clean_string(row['Traditional'])
    pinyin_string = clean_string(row['Pinyin'])
    
    if isinstance(simplified_string, tuple):
        cleaned_dict['Simplified'].append(simplified_string[0])
        cleaned_dict['Simplified'].append(simplified_string[1])
        cleaned_dict['Traditional'].append(traditional_string[0])
        cleaned_dict['Traditional'].append(traditional_string[1])
        cleaned_dict['Pinyin'].append(pinyin_string[0])
        cleaned_dict['Pinyin'].append(pinyin_string[1])
        cleaned_dict['Level'].append(row['Level'])
        cleaned_dict['Level'].append(row['Level'])
    else:
        cleaned_dict['Simplified'].append(simplified_string)
        cleaned_dict['Traditional'].append(traditional_string)
        cleaned_dict['Pinyin'].append(pinyin_string)
        cleaned_dict['Level'].append(row['Level'])
        

# get all 'dirty' strings (strings with unnecessary parentheses or other symbols)
df[df['Simplified'].str.contains('(\（.*\）)|(\|)')].apply(lambda row: split_row(row), axis=1)

df_cleaned = pd.DataFrame(cleaned_dict)
df_cleaned
```

```python
# delete all dirty strings
df = df.drop(df[df['Simplified'].str.contains('(\（.*\）)|(\|)')].index)
```

```python
# add all the cleaned versions of the dirty strings
df = df.append(df_cleaned)
```

```python
# output to csv
df.to_csv('hsk_cleaned.csv', index=False)
```
